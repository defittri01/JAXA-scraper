```python
#  @author Defit Tri H.
# Dowload Raw From Jaxa

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

import os, glob, csv
from xlsxwriter.workbook import Workbook
import pandas as pd
```

```python
chrome_options = Options()
chrome_options.binary_location  = "C:\Program Files\Google\Chrome\Application\chrome.exe"
browser = webdriver.Chrome(chrome_options = chrome_options, executable_path=r"C:\webdriver\chromedriver_win32\chromedriver.exe")
```

Download Raw Data from JAXA

```python
# UID: rainmap
# PW: Niskur+1404

tahun_start = 2001

while True:
    tahun_end   = tahun_start
    url = 'https://sharaku.eorc.jaxa.jp/cgi-bin/trmm/GSMaP/tilemap/show_graph.cgi?flag=2&st={}010100&ed={}123123&lat0=-7.7&lon0=109.6&lat1=-7.7&lon1=109.6&lang=en'.format(str(tahun_start), str(tahun_end))
    browser.get(url)
    browser.execute_script("arguments[0].click();", WebDriverWait(browser, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="graph_dl"]/a[1]'))))
    tahun_start += 1

    if tahun_start == 2022:
        break
```

File Path

```python
files_csv = glob.glob(os.path.join('D:\Kuliah\Semester 4\Sistem Drainase dan Jaringan Air Limbah\Tugas Debit Hujan\JAXA', '*.csv'))
```

Rename csv Files

```python
for file in files_csv:
    start_year = file[104:104+4]
    print(file)
    new_name = 'D:\\Kuliah\\Semester 4\\Sistem Drainase dan Jaringan Air Limbah\\Tugas Debit Hujan\\JAXA\\{} - {}.csv'.format(start_year, str(int(start_year) + 1))
    os.rename(file, new_name)
```

Convert csv to xlsx

```python
# CONVERT CSV to XLSX
for csvfile in files_csv:
    workbook = Workbook(csvfile[:-4] + '.xlsx')
    worksheet = workbook.add_worksheet()
    with open(csvfile, 'rt', encoding='utf8') as f:
        reader = csv.reader(f)
        for r, row in enumerate(reader):
            for c, col in enumerate(row):
                worksheet.write(r, c, col)
    workbook.close()
```
